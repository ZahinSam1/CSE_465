```python
import pygame

# Constants
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600
PREDATOR_RADIUS = 20
FOV_RADIUS = 100  # Field of vision radius
#  is the field of vision radius. It defines how far the 
# predator agent can "see" or sense its surroundings
# set to 100, meaning that the predator can sense objects 
# within a circular area with a radius of 100 pixels around itself. 
# used for visualization purposes. detect object within this range

# Colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))


class Predator:
    def __init__(self, x, y):   #initialize
        self.x = x
        self.y = y

# just visualize the predator and its fov radius
    def draw(self):
        pygame.draw.circle(screen, RED, (int(self.x), int(self.y)), PREDATOR_RADIUS)
        pygame.draw.circle(screen, BLUE, (int(self.x), int(self.y)), FOV_RADIUS, 1)

def main():
    predator = Predator(200, 200)
    # specifying its initial position.

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        screen.fill(WHITE)
        predator.draw()
        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import random
import math

# Constants
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600
PREDATOR_RADIUS = 20
FOV_RADIUS = 100
FOV_ANGLE = 50  # Field of view angle in degrees
WALL_WIDTH = 20
WALL_HEIGHT = 40

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class GameEnv:
    def __init__(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2
        self.wall_x = SCREEN_WIDTH // 3
        self.wall_y = SCREEN_HEIGHT // 3
        self.fov_points = []  # Initialize FOV points

    def move_predator(self, action):
        # Randomly change the predator's position within the screen
        if action == 0:  # Move up
            self.predator_y -= 2
        elif action == 1:  # Move down
            self.predator_y += 2
        elif action == 2:  # Move left
            self.predator_x -= 2
        elif action == 3:  # Move right
            self.predator_x += 2

        # Ensure the predator stays within the screen boundaries
        self.predator_x = max(PREDATOR_RADIUS, min(SCREEN_WIDTH - PREDATOR_RADIUS, self.predator_x))
        self.predator_y = max(PREDATOR_RADIUS, min(SCREEN_HEIGHT - PREDATOR_RADIUS, self.predator_y))

    def draw_environment(self):
        # Draw the wall obstacle
        pygame.draw.rect(screen, BLUE, (self.wall_x, self.wall_y, WALL_WIDTH, WALL_HEIGHT))

        # Draw the predator
        pygame.draw.circle(screen, RED, (int(self.predator_x), int(self.predator_y)), PREDATOR_RADIUS)

        # Draw the FOV (Field of View)
        pygame.draw.polygon(screen, BLUE, self.fov_points, 1)

    def step(self, action):
        # method to move the predator based on the action taken. 
        # It updates the predator's position.
        self.move_predator(action)
        # initializes an empty list fov_points to store the points that define the 
        # Field of View (FOV) polygon.
        # Update FOV points based on the new predator position
        self.fov_points = []
        for angle in range(-FOV_ANGLE // 2, FOV_ANGLE // 2):
            x = self.predator_x + FOV_RADIUS * math.cos(math.radians(angle))
            y = self.predator_y + FOV_RADIUS * math.sin(math.radians(angle))
            self.fov_points.append((int(x), int(y)))

        # dictionary stores the information that will be returned as part of the observation
        observation = {
            # tuple containing the current position of the predator.
            "predator_position": (self.predator_x, self.predator_y),
            # list of points that define the FOV boundary.
            # (x, y) coordinates for each point are added to the fov_points list as a 
            # tuple (int(x), int(y)), where int is used to convert the coordinates to integers.
            "fov_boundary_points": self.fov_points
            # The fov_points list is then used to draw the FOV polygon on the screen, 
            # showing the area that the predator can perceive.
        }
        reward = 0  # good or bad
        done = False  #  This variable is used to determine whether the episode is complete.
        info = {}  # Optional information dictionry
        return observation, reward, done, info

    def reset(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2

        # Initialize FOV points when resetting
        self.fov_points = []
        for angle in range(-FOV_ANGLE // 2, FOV_ANGLE // 2):
            x = self.predator_x + FOV_RADIUS * math.cos(math.radians(angle))
            y = self.predator_y + FOV_RADIUS * math.sin(math.radians(angle))
            self.fov_points.append((int(x), int(y)))

         # Return the initial state as an observation
        observation = {
            "predator_position": (self.predator_x, self.predator_y),
            "fov_boundary_points": self.fov_points
        }
        return observation

def main():
    env = GameEnv()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        action = random.choice([0, 1, 2, 3])
        obs, reward, done, _ = env.step(action)

        screen.fill(WHITE)
        env.draw_environment()
        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import random
import math

# Constants
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600

PREDATOR_RADIUS = 20
FOV_RADIUS = 100
WALL_WIDTH = 20  # Width of the wall
WALL_HEIGHT = 40  # Width of the wall

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class GameEnv:
    def __init__(self):

        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2

        self.wall_x = SCREEN_WIDTH // 3
        self.wall_y = SCREEN_HEIGHT // 3
        self.wall_id = 1  #! Unique ID for the wall

    def move_predator(self, action):
        if action == 0:  # Move up
            self.predator_y -= 2
        elif action == 1:  # Move down
            self.predator_y += 2
        elif action == 2:  # Move left
            self.predator_x -= 2
        elif action == 3:  # Move right
            self.predator_x += 2

        self.predator_x = max(PREDATOR_RADIUS, min(SCREEN_WIDTH - PREDATOR_RADIUS, self.predator_x))
        self.predator_y = max(PREDATOR_RADIUS, min(SCREEN_HEIGHT - PREDATOR_RADIUS, self.predator_y))

    def draw_environment(self):
        pygame.draw.rect(screen, BLUE, (self.wall_x, self.wall_y, WALL_WIDTH, WALL_HEIGHT))
        pygame.draw.circle(screen, RED, (int(self.predator_x), int(self.predator_y)), PREDATOR_RADIUS)
        pygame.draw.circle(screen, BLUE, (int(self.predator_x), int(self.predator_y)), FOV_RADIUS, 1)



    def get_fov_points(self):
        fov_points = {}     #*empty dictionary of fov points
        #* nested loop to iterate over all the x and y coordinates within the FOV radius of the agent's position.
        #* This line iterates over the x-coordinates within the FOV radius
        for x in range(int(self.predator_x - FOV_RADIUS), int(self.predator_x + FOV_RADIUS)):
            #* This line iterates over the y-coordinates within the FOV radius
            for y in range(int(self.predator_y - FOV_RADIUS), int(self.predator_y + FOV_RADIUS)):
                #* This line calculates the distance between the agent's position and the current point (x, y) in the FOV
                distance = math.dist((self.predator_x, self.predator_y), (x, y))
                #* This conditional checks if the calculated distance is less than or equal to the FOV radius
                if distance <= FOV_RADIUS:
                    # *ensures that the point (x, y) is within the screen boundaries
                    if x >= 0 and x < SCREEN_WIDTH and y >= 0 and y < SCREEN_HEIGHT:
                        #* This conditional checks if the point (x, y) is within the boundaries of the wall. 
                        # *If it is, the point is marked with the self.wall_id, indicating that it belongs to the wall.
                        if x >= self.wall_x and x < self.wall_x + WALL_WIDTH and y >= self.wall_y and y < self.wall_y + WALL_HEIGHT:
                            fov_points[(x, y)] = self.wall_id  #! Mark the wall with the wall's unique ID
                            #*If the point is not within the wall boundaries, it is considered empty space and marked with a value of 0
                        else:
                            fov_points[(x, y)] = 0  # Empty space
        # *method returns the fov_points dictionary, 
        #* containing all the FOV points and their respective IDs (wall ID or 0 for empty space)
        return fov_points
    


    def step(self, action):
        self.move_predator(action)
        observation = [self.predator_x, self.predator_y]
        reward = 0
        done = False
        fov_points = self.get_fov_points()
        info = {'wall_id': self.wall_id, 'fov_points': fov_points}
        return observation, reward, done, info

    def reset(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2
        return [self.predator_x, self.predator_y]

def main():

    env = GameEnv()
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
        action = random.choice([0, 1, 2, 3])
        obs, reward, done, info = env.step(action)

        screen.fill(WHITE)
        env.draw_environment()

        # Get FOV points and wall ID from the info dictionary
        fov_points = info['fov_points']
        wall_id = info['wall_id']
        # Loop through FOV points and draw them
        # loop iterates through the items (key-value pairs) in the fov_points dictionary. 
        # Each item consists of a point (coordinates) and its associated ID 
        for point, point_id in fov_points.items():
        # line unpacks the point tuple, separating it into x and y variables, 
        # which represent the coordinates of the FOV point.
            x, y = point
            # conditional checks if the ID of the current FOV point (point_id) matches the ID of the wall (wall_id).
            #  If they match, it means the FOV point is overlapping with the wall.
            if point_id == wall_id:
                pygame.draw.circle(screen, RED, (x, y), 2)  # Draw FOV points overlapping with the wall in red
                print("FOV overlaps with wall at point:", point)
            # else:
            #     pygame.draw.circle(screen, (0, 0, 255, 50), (x, y), 2)  # Draw other FOV points in blue

        pygame.display.flip()
        #! print("Wall ID:", info['wall_id'])  

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import random
import math

# Constants
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600

PREDATOR_RADIUS = 20
FOV_RADIUS = 100
WALLS = {       #dictionary to store wall and its values
    'wall1': {'x': 280, 'y': 200, 'width': 20, 'height': 40, 'id': 1},
    'wall2': {'x': 320, 'y': 190, 'width': 30, 'height': 60, 'id': 2},
    'wall3': {'x': 390, 'y': 250, 'width': 25, 'height': 30, 'id': 3}
}

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class GameEnv:
    def __init__(self):

        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2

    def move_predator(self, action):
        if action == 0:  # Move up
            self.predator_y -= 2
        elif action == 1:  # Move down
            self.predator_y += 2
        elif action == 2:  # Move left
            self.predator_x -= 2
        elif action == 3:  # Move right
            self.predator_x += 2

        self.predator_x = max(PREDATOR_RADIUS, min(SCREEN_WIDTH - PREDATOR_RADIUS, self.predator_x))
        self.predator_y = max(PREDATOR_RADIUS, min(SCREEN_HEIGHT - PREDATOR_RADIUS, self.predator_y))

    def draw_environment(self):
        for wall in WALLS.values():
            pygame.draw.rect(screen, BLUE, (wall['x'], wall['y'], wall['width'], wall['height']))
        pygame.draw.circle(screen, RED, (int(self.predator_x), int(self.predator_y)), PREDATOR_RADIUS)
        pygame.draw.circle(screen, BLUE, (int(self.predator_x), int(self.predator_y)), FOV_RADIUS, 1)

    def get_fov_points(self):
        fov_points = {}  # Dictionary of FOV points
        wall_hit = None  # Variable to track which wall the FOV point hits

        for x in range(int(self.predator_x - FOV_RADIUS), int(self.predator_x + FOV_RADIUS)):
            for y in range(int(self.predator_y - FOV_RADIUS), int(self.predator_y + FOV_RADIUS)):
                distance = math.dist((self.predator_x, self.predator_y), (x, y))
                if distance <= FOV_RADIUS:
                    if x >= 0 and x < SCREEN_WIDTH and y >= 0 and y < SCREEN_HEIGHT:
                        for wall_name, wall_data in WALLS.items():
                            if wall_data['x'] <= x < wall_data['x'] + wall_data['width'] and wall_data['y'] <= y < wall_data['y'] + wall_data['height']:
                                fov_points[(x, y)] = wall_data['id']
                                wall_hit = wall_name  # Store the wall name that the FOV point hits
                                break  # Stop checking other walls for this point
                        else:
                            fov_points[(x, y)] = 0  # Empty space

        return {'fov_points': fov_points, 'wall_hit': wall_hit}
    # *wall_hit: This is a variable used to keep track of which wall, if any, a FOV point hits. It's initially set to None.



    def step(self, action):
        self.move_predator(action)
        observation = {
            'predator_position': [self.predator_x, self.predator_y],
            'fov_points': self.get_fov_points(),
            'walls': WALLS,
        }
        reward = 0
        done = False
        info = {}
        return observation, reward, done, info

    def reset(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2
        observation = {
            'predator_position': [self.predator_x, self.predator_y],
            'fov_points': self.get_fov_points(),
            'walls': WALLS,
        }
        return observation


    


def main():
    env = GameEnv()
    running = True

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        action = random.choice([0, 1, 2, 3])
        observation, reward, done, info = env.step(action)

        screen.fill(WHITE)
        env.draw_environment()

        # Access the predator's position, FOV points, and wall information
        predator_position = observation['predator_position']
        fov_points = observation['fov_points']
        walls_info = observation['walls']
        wall_hit = observation.get('wall_hit')



        # Loop through FOV points and draw them
        for point, point_id in fov_points.items():
            if point_id:  # Check if the point overlaps with a wall
                if len(point) == 2:  # Check if point is a valid (x, y) tuple
                    x, y = point  # Unpack the (x, y) coordinates from the dictionary key (point)
                    pygame.draw.circle(screen, RED, (x, y), 2)  # Draw FOV points overlapping with a wall in red
                    wall_names = [wall_name for wall_name, wall_data in WALLS.items() if wall_data['id'] == point_id]
                    if wall_names:
                        print(f"FOV overlaps with {', '.join(wall_names)} at point: ({x}, {y})")
                    else:
                        print(f"Invalid point: {point}")

                
        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()

```

```python
import pygame
import random
import math

# Constants
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600

PREDATOR_RADIUS = 20
FOV_RADIUS = 100
WALLS = {  # Dictionary to store walls and their values
    'wall1': {'x': 280, 'y': 200, 'width': 20, 'height': 40, 'id': 1},
    'wall2': {'x': 320, 'y': 190, 'width': 30, 'height': 60, 'id': 2},
    'wall3': {'x': 390, 'y': 250, 'width': 25, 'height': 30, 'id': 3}
}

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class GameEnv:
    def __init__(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2

    def move_predator(self, action):
        if action == 0:  # Move up
            self.predator_y -= 2
        elif action == 1:  # Move down
            self.predator_y += 2
        elif action == 2:  # Move left
            self.predator_x -= 2
        elif action == 3:  # Move right
            self.predator_x += 2

        self.predator_x = max(PREDATOR_RADIUS, min(SCREEN_WIDTH - PREDATOR_RADIUS, self.predator_x))
        self.predator_y = max(PREDATOR_RADIUS, min(SCREEN_HEIGHT - PREDATOR_RADIUS, self.predator_y))

    def draw_environment(self):
        for wall in WALLS.values():
            pygame.draw.rect(screen, BLUE, (wall['x'], wall['y'], wall['width'], wall['height']))
        pygame.draw.circle(screen, RED, (int(self.predator_x), int(self.predator_y)), PREDATOR_RADIUS)
        pygame.draw.circle(screen, BLUE, (int(self.predator_x), int(self.predator_y)), FOV_RADIUS, 1)

    def get_fov_points(self):
        fov_points = {}  # Dictionary of FOV points

        for x in range(int(self.predator_x - FOV_RADIUS), int(self.predator_x + FOV_RADIUS)):
            for y in range(int(self.predator_y - FOV_RADIUS), int(self.predator_y + FOV_RADIUS)):
                if x >= 0 and x < SCREEN_WIDTH and y >= 0 and y < SCREEN_HEIGHT:
                    fov_points[(x, y)] = 0  # Set FOV points as 0 (empty space)

        for wall_name, wall_data in WALLS.items():
            for x in range(wall_data['x'], wall_data['x'] + wall_data['width']):
                for y in range(wall_data['y'], wall_data['y'] + wall_data['height']):
                    fov_points[(x, y)] = wall_data['id']  # Set FOV points in walls as the wall's ID

        return fov_points

    def step(self, action):
        self.move_predator(action)
        observation = {
            'predator_position': [self.predator_x, self.predator_y],
            'fov_points': self.get_fov_points(),
            'walls': WALLS,
        }
        reward = 0
        done = False
        info = {}

        # Check for FOV overlaps with walls and print the result
        fov_points = observation['fov_points']
        for (x, y), wall_id in fov_points.items():
            wall_id = fov_points[(x, y)]  # Get the wall ID directly from fov_points
            if wall_id in WALLS.values():  # Check if FOV overlaps with a wall
                wall_name = [name for name, data in WALLS.items() if data['id'] == wall_id]
                if wall_name:
                    print(f"FOV overlaps with wall '{wall_name[2]}' at point: ({x}, {y})")

        return observation, reward, done, _, info

    def reset(self):
        self.predator_x = SCREEN_WIDTH // 2
        self.predator_y = SCREEN_HEIGHT // 2
        observation = {
            'predator_position': [self.predator_x, self.predator_y],
            'fov_points': self.get_fov_points(),
            'walls': WALLS,
        }
        return observation

def main():
    env = GameEnv()
    running = True

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        action = random.choice([0, 1, 2, 3])
        observation, reward, done, info, _ = env.step(action)

        screen.fill(WHITE)
        env.draw_environment()

        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```
