```python
pos1 = np.array(
        [np.random.uniform(30, 400 - 30), np.random.uniform(30, 400 - 30)], 
        dtype=np.float32
        )

pos2 = np.array(
        [np.random.uniform(30, 400 - 30), np.random.uniform(30, 400 - 30)], 
        dtype=np.float32
        )

print(pos1, pos2)
```

```python
direction = pos2 - pos1
distance_between_centers = np.linalg.norm(direction)
# current_distance = distance_between_centers
print(direction)
```

```python
import gym
from gym import spaces      #spaces to define observation spaces
import numpy as np

class ObsENV(gym.Env):      #custom env class that inherits from gym
    def __init__(self):         #initalizing the environment
        super(ObsENV, self).__init__()

        # creating a list of obstacle categories with 3 values
        self.ObsENV_categories = ["Fire", "Bullets", "Walls"]
        # calculating the number of list using len
        self.num_categories = len(self.ObsENV_categories)

        # made a tuple(list of legit anything but ordered) called shape that represents the dimensions of the 
        # one hot encoded observation. 1d array with length equal to number of obstacle categories
        shape = (self.num_categories,)

        # two numpy arrays called low and high to define range of one hot encoding
        # low means all zeros, and high means all ones.
        # they represent the minimum and maximum values for each element in one hot encoded observation
        low = np.zeros(shape, dtype=np.float32)
        high = np.ones(shape, dtype=np.float32)

        # defining observation space using space.box. its a continuous observation space
        # low and high value define range for each element in one hot encoded observation
        self.observation_space = spaces.Box(low, high, dtype=np.float32)

    def reset(self):
        # Generate a random obstacle category like fire, bullet or walls
        random_obstacle = np.random.choice(self.ObsENV_categories)
        
        # Create a one-hot encoded observation by initializing an array of zeros and setting 
        # the element corresponding to the selected obstacle category to 1
        # This observation represents the initial obstacle condition at the start of the episode.
        observation = np.zeros(self.num_categories)
        observation[self.ObsENV_categories.index(random_obstacle)] = 1.0
        
        return observation

    def step(self, action):
        # Define your custom step logic here
        pass

# Create and use the custom Weather environment
env = ObsENV()
obs = env.reset()
print("One-hot encoded obstacle observation:", obs)
```
